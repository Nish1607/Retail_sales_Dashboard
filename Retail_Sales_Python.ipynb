```python
from sqlalchemy import create_engine
import pandas as pd

# Set up connection string for Windows Authentication
server = 'NISHISONI\\SQLEXPRESS'
database = 'Retail_sales'
driver = 'ODBC Driver 17 for SQL Server'

# SQLAlchemy connection string
connection_string = f"mssql+pyodbc://@{server}/{database}?driver={driver.replace(' ', '+')}&trusted_connection=yes"

# Create engine
engine = create_engine(connection_string)

# Query your table
query_Products = "SELECT * FROM Products_Staging"
query_Dates= "SELECT * FROM Date_Staging"
query_Customers= "SELECT  * FROM Customers_Staging"
query_Sales= "SELECT  * FROM Sales_Staging"
query_Stores= "SELECT * FROM Stores_Staging"

# Read both tables into separate DataFrames
df_products = pd.read_sql(query_Products, engine)
df_dates = pd.read_sql(query_Dates, engine)
df_customers = pd.read_sql(query_Customers, engine)
df_Sales = pd.read_sql(query_Sales, engine)
df_Stores = pd.read_sql(query_Stores, engine)
# Show both results
print("Products Table:")
print(df_products.head())
print("\nDates Table:")
print(df_dates.head())
print("\nCustomers Table:")
print(df_customers.head())
print("\nSales Table:")
print(df_Sales.head())
print("\nStores Table:")
print(df_Stores.head())
```

```python
# Add a new column for better readability
df_Sales['DiscountPercentage'] = (df_Sales['CleanedDiscountPct'] * 100).round(2)
```

```python
def discount_category(pct):
    if pct >= 30:
        return 'High Discount'
    elif pct >= 10:
        return 'Medium Discount'
    elif pct > 0:
        return 'Low Discount'
    else:
        return 'No Discount'

df_Sales['DiscountCategory'] = df_Sales['DiscountPercentage'].apply(discount_category)
```
